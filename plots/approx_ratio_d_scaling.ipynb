```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, subplot
from numpy import arange, array, sqrt
import json
```

```python
dat = json.load(open('../angles_regular_graphs.json'))
```

```python
fig = figure(figsize=(6,8))
ARs = {q:[] for q in range(1,12)}
farhi = [0.3033,0.4075,0.4726,0.5157,0.5476,0.5721,0.5915]
for dd in range(3,12):
    for q in range(1,12):
        if str(q) in dat[str(dd)]:
            ARs[q].append(dat[str(dd)][str(q)]['AR'])
ax1 = subplot(2,1,1)
ax2 = subplot(2,1,2)
for q in range(1,12):
    dd_ = arange(3,12)[0:len(ARs[q])]
    ax1.plot(dd_,ARs[q],'o-',label='p='+str(q))
    ax2.plot(dd_,(array(ARs[q])-0.5)*sqrt(dd_),'o-')
for q in range(5):
    ax2.plot([2.5,11.5],2*[farhi[q]],'C'+str(q)+'--')
ax2.set_xlabel('Connectivity d')
ax1.set_ylabel('Expectation value')
ax2.set_ylabel('Coefficient V(D)')
#ax1.legend()
ax1.set_xticklabels([])
ax2.plot([2.75,11.25],[0.763166,0.763166],'r--')
ax1.axis([2.75,11.25,0.55,0.9])
ax2.axis([2.75,11.25,0.25,0.8])
```
