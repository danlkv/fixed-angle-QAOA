```python
import matplotlib.pyplot as plt
import qtensor

plt.rcParams.update({'font.size': 16})
```

```python
SELECT_P = [3, 6, 11]
```

```python
def get_colors(num=2, offset=0.25, cmap=plt.cm.gnuplot2):
    """
    Returns:
        Array of len ``num``.
    """
    return np.array([cmap(x) for x in np.linspace(offset, 1-offset, num)])
```

```python
def plot_gamma_beta(gamma, beta):
    color_gamma, color_beta = get_colors()
    plt.plot(gamma, 'o-', label=r'$\gamma$', color=color_gamma)
    plt.plot(beta, 'o-', label=r'$\beta$', color=color_beta)
    plt.xticks(range(len(gamma)))
```

```python
fig, axs = plt.subplots(1, 3, figsize=(8*2, 8*2/3 -1))
axs = iter(axs)

for p in SELECT_P:
    plt.sca(next(axs))
    gb = qtensor.tools.BETHE_QAOA_VALUES[str(p)]['angles']
    gamma, beta = gb[:p], gb[p:]
    plot_gamma_beta(-2*np.array(gamma), beta)
    plt.title(f'p={p}')
    plt.grid(alpha=0.3)
    plt.xlabel('Index')
    
plt.gca().legend(bbox_to_anchor=(1.02, 1.05))
plt.sca(fig.axes[0])
plt.ylabel('Angle')
plt.tight_layout()
plt.savefig('./pdf/gamma_beta_p3p6p11.pdf')
```
